```python
import yfinance as yf
import matplotlib.pyplot as plt
plt.style.use("seaborn")
import warnings
warnings.simplefilter("ignore")
from datetime import datetime as dt
tickers = yf.Tickers('eth btc msft')


import pandas as pd
msft = yf.Ticker("MSFT")
history = msft.history(period="20y")
history.index = pd.to_datetime(history.index)
history = history.resample("Y")["Dividends"].sum()
history = history.reset_index()
history["Year"] = history["Date"].dt.year

plt.figure(figsize=(10,6))
ax = plt.subplot()
ax.bar(x=history.Year, height=history.Dividends)
ax.bar_label(ax.containers[0])
ax.set(title="Microsoft Dividends over 20 years",
       xlabel="Year",
       ylabel="Dividends")
plt.show()
```
